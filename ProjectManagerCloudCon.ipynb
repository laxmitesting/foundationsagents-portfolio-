```python
import os
import json
import pandas as pd
from openai import OpenAI
from IPython.display import display, Markdown
from dotenv import load_dotenv
```

```python
load_dotenv(override=True)
```

```python
# Set Pandas options to show full paragraph content natively in your HTML view
pd.set_option("display.max_colwidth", None)
```

```python
# =====================================================================
# 1. INITIALIZE ALL CLIENTS USING THE OPENAI LIBRARY
# =====================================================================
openai_client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

gemini_client = OpenAI(
    api_key=os.getenv("GOOGLE_API_KEY"),
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)

cerebras_client = OpenAI(
    api_key=os.getenv("CEREBRAS_API_KEY"),
    base_url="https://api.cerebras.ai/v1"
)
```

```python
# =====================================================================
# 2. THE BUSINESS CRISIS INPUT DATA (Cloud Migration Framework)
# =====================================================================
sprint_backlog_incident = """
PROJECT METRICS:
- Project Name: Legacy App GCP Migration
- Current Status: Delayed (Migration window closes in 5 days; blockades flagged)
- Team Status: Blocked by conflicting migration priorities

BACKLOG ITEMS VYEING FOR PRIORITY:
1. Lift-and-Shift Task A: 'Migrate Front-End Web Servers' (Move the user interface as-is to the cloud. Strictly fulfills the immediate executive deadline, but copies old, messy configuration).
2. Security & IAM Task B: 'Fix Multi-Factor Authentication & Least-Privilege Access' (The security team is warning that root access keys are currently shared in the test environment; critical deployment risk).
3. Cloud Optimization Task C: 'Refactor Database to Native Cloud RDS' (Rewriting the database layer to utilize cloud-native backups and auto-scaling. The code is ready but requires team time to validate).
"""

user_prompt = f"Review the current cloud migration backlog and justify what must be prioritized:\n\n{sprint_backlog_incident}"
```

```python
# =====================================================================
# 3. RUN THE PANEL IN PARALLEL
# =====================================================================
print("--- Gathering Expert Perspectives ---")

# Agent 1: OpenAI as the Aggressive PM
print("Polling OpenAI (Product Manager Persona)...")
op_res = openai_client.chat.completions.create(
    model="gpt-4o",
    messages=[
        {"role": "system", "content": "You are a hyper-focused, revenue-driven Senior Product Manager. Prioritize things purely based on competitive speed-to-market, retention metrics, user growth, and near-term ARR impact. Be direct and uncompromising about shipping features."},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.3
)
openai_text = op_res.choices[0].message.content
```

```python
# Agent 2: Gemini as the Systems Lead
print("Polling Gemini (Systems Architect Persona)...")
gem_res = gemini_client.chat.completions.create(
    model="gemini-2.5-flash-lite",
    messages=[
        {"role": "system", "content": "You are a Principal Software Architect. Prioritize things purely based on codebase resilience, reduction of system fragility, preventing system downtime cascades, and removing developer friction. Explicitly call out product decisions that ignore core architecture scaling problems."},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.4
)
gemini_text = gem_res.choices[0].message.content
```

```python
# Agent 3: Cerebras as the Scrum Master
print("Polling Cerebras (Agile Delivery Coach Persona)...")
cer_res = cerebras_client.chat.completions.create(
    model="gpt-oss-120b",
    messages=[
        {"role": "system", "content": "You are an expert Agile Delivery Coach and Scrum Master. Prioritize items purely based on restoring consistent team velocity, resolving burnout markers, establishing reliable sprint completion statistics, and defusing inter-team friction."},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.2
)
cerebras_text = cer_res.choices[0].message.content

print("\nAll panels successfully calculated outputs.")
```

```python
# =====================================================================
# 4. ORGANIZE AND DISPLAY THE MATRIX VIEW (Updated for Cloud Migration)
# =====================================================================
arbitrator_data = {
    "Analysis Vector": [
        "Role Persona Focus", 
        "Assessment of Current Delay", 
        "Hard Strategic Recommendation"
    ],
    "OpenAI (Product Manager)": [
        "Timeline Compliance & Target Deadlines",
        "Missing the 5-day migration window poses a massive reputational and operational threat to corporate timelines.",
        "Force execution of Lift-and-Shift Task A immediately. Fulfill the deadline first; we can clean up the configuration clutter after the servers are safely live in GCP."
    ],
    "Gemini (Systems Lead)": [
        "Infrastructure Resilience & Cloud Architecture",
        "Moving unoptimized, messy legacy configurations directly to the cloud creates massive systemic technical debt from day one.",
        "Halt the blind lift-and-shift. Prioritize Task C (RDS Refactoring) to ensure our database architecture is stable, scalable, and cloud-native before moving production traffic."
    ],
    "Cerebras (Scrum Master)": [
        "Operational Security & Team Guardrails",
        "Shared root access keys in the test environment represent a critical compromise of team deployment safety and governance.",
        "Pause both the deadline rush and the database optimization. Execute Security Task B immediately to establish mandatory IAM guardrails before any cloud platform changes go live."
    ]
}

df = pd.DataFrame(arbitrator_data)
df.set_index("Analysis Vector", inplace=True)

print("\n" + "="*40)
print("CLOUD MIGRATION ARBITRATION MATRIX VIEW")
print("="*40)
display(df)
```

```python
# =====================================================================
# 5. THE MASTER JUDGE (Using Zero-Installation HTML String Parsing)
# =====================================================================
print("\n--- Submitting Cloud Migration Matrix to the VP of Engineering (Master Judge) ---")

# Convert your dataframe directly to a raw HTML table string (No dependencies needed!)
matrix_as_html = df.to_html()

judge_system_prompt = """
You are the Vice President of Engineering. You are reviewing an HTML table summarizing 
the conflicting priorities of your Product Manager (deadline driven), Principal Architect (architecture driven), 
and Scrum Master (security/process driven) regarding a critical GCP migration that is 5 days from closing.

You must act as the executive tie-breaker. Evaluate their perspectives, weigh the risk of missing the 5-day 
deadline against the risk of launching with shared root access keys or sloppy configurations, and issue a unified directive.

Output your final tactical directive in a strict JSON object using these exact keys:
{
  "clash_synthesis": "A brief summary analyzing the tension between meeting the corporate deadline versus maintaining security and architecture standards.",
  "executive_verdict": "Your explicit compromise decision on what gets deployed first during this 5-day window.",
  "ordered_migration_steps": [
     "Step-by-step sequence of how Task A, Task B, and Task C must be executed or staged over the next 5 days."
  ],
  "stakeholder_alignment_narrative": "How you will message this decision to the team to ensure the PM, Architect, and Security teams execute cohesively."
}
"""

judge_response = openai_client.chat.completions.create(
    model="gpt-4o",
    messages=[
        {"role": "system", "content": judge_system_prompt},
        {"role": "user", "content": f"Review this stakeholder migration matrix HTML:\n\n{matrix_as_html}"}
    ],
    response_format={"type": "json_object"},
    temperature=0.3
)

# Parse and display the final JSON response cleanly
final_arbitration = json.loads(judge_response.choices[0].message.content)

print("\n" + "="*60)
print("FINAL EXECUTIVE CLOUD MIGRATION DIRECTIVE")
print("="*60)
print(json.dumps(final_arbitration, indent=2))


# =====================================================================
# 6. RENDER THE EXECUTIVE DIRECTIVE AS BEAUTIFUL HUMAN MARKDOWN
# =====================================================================
# Parse and display the final JSON response cleanly
final_arbitration = json.loads(judge_response.choices[0].message.content)

# Construct a clean Markdown string from the JSON keys
markdown_output = f"""
## 🏛️ Executive VP of Engineering Directive

### 🎯 Strategic Conflict Synthesis
{final_arbitration['clash_synthesis']}

---

### ⚖️ The Verdict
**{final_arbitration['executive_verdict']}**

---

### 📅 Ordered 5-Day Migration Roadmap
"""

for step in final_arbitration['ordered_migration_steps']:
    markdown_output += f"* **{step.split(':')[0]}:** {step.split(':')[1].strip()}\n"

markdown_output += f"""
---

### 📢 Stakeholder Alignment Narrative
*{final_arbitration['stakeholder_alignment_narrative']}*
"""

# Render it natively inside Jupyter
display(Markdown(markdown_output))
```
