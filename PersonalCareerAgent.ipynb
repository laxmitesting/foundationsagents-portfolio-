````python
import os
from openai import OpenAI
from IPython.display import display, Markdown

# Initialize endpoints using the unified API protocol
openai_client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
cerebras_client = OpenAI(
    api_key=os.getenv("CEREBRAS_API_KEY"),
    base_url="https://api.cerebras.ai/v1"
)

# =====================================================================
# 1. INPUT: YOUR RAW EXPERIENCE DRAFT
# =====================================================================
# Change this text to any raw project description from your actual work!
raw_experience = """
I spent a couple of weeks completing a Cloud Platform AI Engineering course. 
The course took me through concepts of using AI to build CI/CD pipelines, 
create infrastructure templates and use AI to debug security logs. 
I had to complete labs to successfully get the completion certification. 
These labs were fairly time-consuming and challenging. 
Even with the use of AI, the work did not feel automatic and I 
learned that using AI for coding requires critical thinking and patience. 
"""

# =====================================================================
# 2. ARCHITECTURE: EVALUATOR-OPTIMIZER LOOP
# =====================================================================
print("--- Launching Personal Career Dashboard: Resume Auditor ---\n")

initial_prompt = f"""
Transform this raw experience description into a highly impactful, single-sentence resume bullet point.
Raw Experience: {raw_experience}
"""

print("Drafting Initial Bullet Point (Cerebras gpt-oss-120b)...")
gen_res = cerebras_client.chat.completions.create(
    model="gpt-oss-120b",
    messages=[
        {
            "role": "system", 
            "content": "You are an expert Technical Resume Writer. Craft a single, powerful resume bullet point using direct, engineering-centric language. Do not include introductory text."
        },
        {"role": "user", "content": initial_prompt}
    ],
    temperature=0.3
)
current_bullet = gen_res.choices[0].message.content
print(f"-> Initial Draft Created.\n")

# Start the evaluation loops
for turn in range(1, 3):
    print(f"🔄 Quality Audit Turn {turn}...")
    
    audit_prompt = f"""
    Review this resume bullet point against top-tier tech industry standards.
    Bullet Point: {current_bullet}
    """
    
    eval_res = openai_client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {
                "role": "system", 
                "content": (
                    "You are a critical Technical Hiring Manager. Audit the resume bullet point. "
                    "CRITICAL RULES:\n"
                    "1. Ban all fluffy action verbs (e.g., helped, assisted, managed, spearheaded).\n"
                    "2. Mandate direct, engineering-centric terminology like 'Implementation', 'Remediation', or 'Configuration'.\n"
                    "3. Ensure there is a clear technical action and an implied operational outcome.\n"
                    "If the bullet point perfectly meets these criteria, reply with ONLY the word 'APPROVED'."
                )
            },
            {"role": "user", "content": audit_prompt}
        ],
        temperature=0.1
    )
    feedback = eval_res.choices[0].message.content
    print(f"   [Manager Feedback]: {feedback}\n")
    
    if "APPROVED" in feedback.upper():
        print("✅ Bullet point cleared the quality gate! Finalizing.")
        break
        
    # Optimizer step
    optimizer_prompt = f"Rewrite this resume bullet point to fully satisfy the feedback. Original: {current_bullet}\nFeedback: {feedback}"
    
    gen_res = cerebras_client.chat.completions.create(
        model="gpt-oss-120b",
        messages=[
            {
                "role": "system", 
                "content": "You are an expert Technical Resume Writer. Revise the single bullet point based on the feedback. Maintain a strict, direct, engineering style. Do not write an intro or conversational filler."
            },
            {"role": "user", "content": optimizer_prompt}
        ],
        temperature=0.1
    )
    current_bullet = gen_res.choices[0].message.content
    print(f"   [Optimized Bullet {turn}]: {current_bullet}\n")

# =====================================================================
# 3. DISPLAY FINAL RESULT
# =====================================================================
print("\n" + "="*50)
print("LAB OUTCOME: RESUME AUDIT COMPLETE")
print("="*50)

output_report = f"""
# 📈 Personal Career Dashboard: Resume Optimization
**Architecture Pattern:** Evaluator-Optimizer

---

### 📝 Raw Input Experience
*"{raw_experience.strip()}"*

---

### 🚀 Production-Ready Resume Bullet (Approved)
* **{current_bullet.strip().lstrip('* ')}**

---
*Note: Evaluated for direct engineering impact by OpenAI gpt-4o and optimized by Cerebras gpt-oss-120b.*
"""

display(Markdown(output_report))
````

🗺️ Pattern 2: Orchestrator-Workers (The Dynamic Career Pivot Blueprint)
If you want to transition into a new technical domain (e.g., moving from pure recruiting or traditional coding into designing human-centered AI systems), a single prompt cannot give you an actionable roadmap. This is where the Orchestrator-Workers pattern shines.

You feed the system a high-level goal: "Create a comprehensive 90-day transition blueprint to shift my professional profile into a specialized AI Engineering Lead."

The Orchestrator (gpt-4o): Analyzes the gap between your current profile and the target industry standard. It breaks the 90-day plan down into discrete, specialized assignments.

Worker 1: The Curriculum Designer (Gemini 1.5 Pro): Receives an assignment to map out a precise academic reading and course tracking list.

Worker 2: The Portfolio Architect (Cerebras gpt-oss-120b): Receives an assignment to design three specific, non-trivial open-source side projects that demonstrate production-grade AI pipeline execution.

Worker 3: The Market Positioner (gpt-4o): Receives an assignment to construct a targeted networking strategy, content creation schedule, and LinkedIn brand optimization layout.

Why this fits the pattern:
The workers have no idea what the overall 90-day strategy looks like until the Orchestrator dynamically generates their specific task assignments based on your background. The final output is aggregated into a vertically stacked, comprehensive master playbook.

````python
import json
import os
from openai import OpenAI
from IPython.display import display, Markdown

# Initialize all endpoints using the unified API protocol
openai_client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
gemini_client = OpenAI(
    api_key=os.getenv("GOOGLE_API_KEY"),
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)
cerebras_client = OpenAI(
    api_key=os.getenv("CEREBRAS_API_KEY"),
    base_url="https://api.cerebras.ai/v1"
)

# =====================================================================
# 1. INPUT: YOUR STRATEGIC CAREER PIVOT GOAL
# =====================================================================
pivot_goal = """
I want to transition my professional profile over the next 90 days to focus on 
human-centered Artificial Intelligence systems. I have a background 
as a Business Psychologist, Support Engineer and Technical Account Manager. 
I am now currently upskilling with the Agentic AI course. 
I would like my career pivot to blend the technology and psychology skills. 
"""

# =====================================================================
# 2. ARCHITECTURE: ORCHESTRATOR GENERATES SPECIALIZED WORK TASKS (REALIGNED)
# =====================================================================
print("--- Launching Personal Career Dashboard: Orchestrator-Workers ---")
print("Orchestrator (gpt-4o) is breaking down the 90-day strategy into tasks...")

orchestrator_prompt = f"""
Analyze this 90-day professional pivot goal:
{pivot_goal}

Deconstruct this goal into exactly two distinct, highly specialized operational assignments:
- 'task_for_translation_worker': Focus on market positioning, mapping complex engineering concepts to business value, identifying high-demand industry keywords, and building communication frameworks.
- 'task_for_portfolio_worker': Focus on designing a practical, multi-agent software engineering project that serves as a public portfolio piece.

CRITICAL RETURN INSTRUCTION:
Return your response as a raw, valid JSON object. 
Do NOT wrap your response in markdown code blocks like ```json ... ```. 
Use these exact keys: 'task_for_translation_worker' and 'task_for_portfolio_worker'.
"""

orch_res = openai_client.chat.completions.create(
    model="gpt-4o",
    messages=[
        {"role": "system", "content": "You are an Elite Executive Career Strategist. You excel at breaking down broad goals into highly technical, separate execution assignments."},
        {"role": "user", "content": orchestrator_prompt}
    ],
    response_format={"type": "json_object"}
)

# Parse the dynamic assignments out of the orchestrator's response
assignments = json.loads(orch_res.choices[0].message.content)

# =====================================================================
# 3. DYNAMICALLY DEPLOY THE WORKERS WITH STRINGIFIED PAYLOADS (FIXED)
# =====================================================================

# FIX: If gpt-4o nested a dictionary inside the key, flatten it back to a clean string format
gemini_task_payload = assignments['task_for_translation_worker']
if isinstance(gemini_task_payload, dict):
    gemini_task_payload = json.dumps(gemini_task_payload, indent=2)

cerebras_task_payload = assignments['task_for_portfolio_worker']
if isinstance(cerebras_task_payload, dict):
    cerebras_task_payload = json.dumps(cerebras_task_payload, indent=2)


print(" -> Deploying Worker 1 (Gemini 2.5 Flash Lite) as Technical Translation Recruiter...")
worker_1_res = gemini_client.chat.completions.create(
    model="gemini-2.5-flash-lite",  
    messages=[
        {
            "role": "system", 
            "content": (
                "You are a Principal Technical Recruiter and Talent Acquisition Director who bridges "
                "the gap between complex engineering concepts and non-technical business partners. "
                "Your job is to translate the assigned technical data structure into a precise 90-day roadmap "
                "for building professional bridges, identifying key industry keywords, mastering technical "
                "vocabulary, and learning how to explain advanced AI workflows to both corporate executives and engineers."
            )
        },
        {"role": "user", "content": gemini_task_payload} # <-- FIXED: Passing sanitized string payload
    ]
)

print(" -> Deploying Worker 2 (Cerebras gpt-oss-120b) on Portfolio Architecture...")
worker_2_res = cerebras_client.chat.completions.create(
    model="gpt-oss-120b",
    messages=[
        {
            "role": "system", 
            "content": "You are a Principal AI Software Engineer. Design a concrete, non-trivial, multi-agent technical project architecture that the user can build to prove their engineering skills. Keep your response direct and structured."
        },
        {"role": "user", "content": cerebras_task_payload} # <-- FIXED: Passing sanitized string payload
    ],
    temperature=0.2
)

print("\nAll career development tracks generated successfully.")

# =====================================================================
# 4. RENDER THE VERTICALLY STACKED BLUEPRINT
# =====================================================================
print("Compiling your master career playbook...")

translation_task = assignments['task_for_translation_worker'] # <-- Fixed: Realigned variable capture
portfolio_task = assignments['task_for_portfolio_worker']
translation_output = worker_1_res.choices[0].message.content
portfolio_output = worker_2_res.choices[0].message.content

master_playbook = f"""
# 🗺️ 90-Day Strategic Career Pivot Playbook
**Architecture Pattern:** Orchestrator-Workers Pattern (Fully Realigned)

---

## 🧠 1. Strategic Breakdown by the Incident Commander (Orchestrator)
The central strategist analyzed your career pivot and generated two parallel tracks:

### 📢 Track Alpha: Technical Vocabulary & Translation Strategy
* **Assigned Worker:** Gemini 2.5 Flash Lite (Technical Translation Recruiter)
* **Directive Issued:** > {translation_task}

### 💻 Track Bravo: Technical Portfolio Execution
* **Assigned Worker:** Cerebras gpt-oss-120b (Principal Architect)
* **Directive Issued:** > {portfolio_task}

---

## 🛠️ 2. Tailored Operational Execution Roadmaps

### 🔴 Market Positioning, Keywords & Communication Syllabus (From Gemini)
{translation_output}

---

### 🔵 Production Capstone Project Blueprint (From Cerebras)
{portfolio_output}

---
**End of Blueprint. Your custom translation and execution tracks are live.**
"""

print("\n" + "="*50)
print("DYNAMIC TRANSFORMATION BLUEPRINT OUTPUT")
print("="*50)
display(Markdown(master_playbook))
````

````python
import os
from openai import OpenAI
from IPython.display import display, Markdown

# Initialize all endpoints using the unified API protocol
openai_client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
gemini_client = OpenAI(
    api_key=os.getenv("GOOGLE_API_KEY"),
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)
cerebras_client = OpenAI(
    api_key=os.getenv("CEREBRAS_API_KEY"),
    base_url="https://api.cerebras.ai/v1"
)

# =====================================================================
# 1. INPUT: HYBRID GCP ARCHITECT / TAM PROFILE
# =====================================================================
target_experience = """
Certified Google Cloud (GCP) Professional Cloud Architect with a proven track record 
designing resilient, production-grade cloud infrastructure while simultaneously operating 
as a Technical Account Manager (TAM). Expert at translating architecture roadmaps 
into engineering milestones, managing high-stakes incident response remediation, and 
structuring operational governance matrices to maximize infrastructure ROI for clients.
"""

# =====================================================================
# 2. ARCHITECTURE: PARALLEL EXECUTION (SIMULTANEOUS API CALLS)
# =====================================================================
print("--- Launching Personal Career Dashboard: Parallel Mock Panel ---")
print("Firing concurrent API requests to all interview panelists based on hybrid GCP/TAM profile...\n")

````

````python
# Panelist 1: The Recruiter (Gemini)
print(" -> Panelist 1 (Gemini 2.5 Flash Lite) is analyzing communication & TAM metrics...")
prompt_recruiter = f"Analyze this hybrid GCP Architect/TAM profile and generate the single most critical interview question regarding business translation, client retention metrics, or commercial ROI keyword alignment: {target_experience}"

res_recruiter = gemini_client.chat.completions.create(
    model="gemini-2.5-flash-lite",
    messages=[
        {"role": "system", "content": "You are a Principal Technical Recruiter screening for a high-value TAM role. Ask one direct, high-impact behavioral or business-value interview question. Do not include intro greetings or explanations."},
        {"role": "user", "content": prompt_recruiter}  # <-- FIXED: Handed the prompt string here!
    ],
    max_tokens=200
)

# Panelist 2: The Systems Architect (Cerebras)
print(" -> Panelist 2 (Cerebras gpt-oss-120b) is analyzing GCP infrastructure depth...")
prompt_architect = f"Analyze this hybrid GCP Architect/TAM profile and generate the single most punishing technical architecture question regarding GCP-specific native services, scale bottlenecks, or multi-region failovers: {target_experience}"

res_architect = cerebras_client.chat.completions.create(
    model="gpt-oss-120b",
    messages=[
        {"role": "system", "content": "You are a Principal Systems Architect checking for real engineering depth behind the certification. Ask one highly technical GCP-specific architecture interview question. Get straight to the infrastructure core."},
        {"role": "user", "content": prompt_architect} # <-- Verified payload pass
    ],
    temperature=0.2,
    max_tokens=700
)

# Panelist 3: The Security Auditor (OpenAI)
print(" -> Panelist 3 (OpenAI gpt-4o) is analyzing operational governance & risk...")
prompt_security = f"Analyze this hybrid GCP Architect/TAM profile and generate the single most critical interview question regarding enterprise security boundaries, IAM hierarchy governance, compliance matrices, or responsibility boundaries during cloud incidents: {target_experience}"

res_security = openai_client.chat.completions.create(
    model="gpt-4o",
    messages=[
        {"role": "system", "content": "You are a Senior Security Legal Auditor checking for enterprise operational risk management. Ask one sharp interview question about security governance boundaries or compliance containment."},
        {"role": "user", "content": prompt_security} # <-- Verified payload pass
    ],
    temperature=0.2,
    max_tokens=200
)

print("\nAll panelists have submitted their evaluations independently.")

````

````python
# =====================================================================
# 3. COLLECT AND DISPLAY THE REVIEW PANEL
# =====================================================================
print("Compiling panel review matrix...")

q_recruiter = res_recruiter.choices[0].message.content.strip()
q_architect = res_architect.choices[0].message.content.strip()
q_security = res_security.choices[0].message.content.strip()

panel_report = f"""
# 🎤 360° Parallel Mock Interview Review
**Target Hybrid Profile Evaluated:**
> *{target_experience.strip()}*

---

### 📢 Panelist 1: The Technical Translation Recruiter (Gemini)
* **Focus Area:** Business Translation, Client Health, & TAM Keyword Alignment  
* **Interview Question:** > {q_recruiter}

---

### 💻 Panelist 2: The Principal Systems Architect (Cerebras)
* **Focus Area:** Core GCP Infrastructure, Scale, & Engineering Proof  
* **Interview Question:** > {q_architect}

---

### 🛡️ Panelist 3: The Senior Security Auditor (OpenAI)
* **Focus Area:** Identity Access Management (IAM), Compliance Governance, & Incident Boundaries  
* **Interview Question:** > {q_security}

---
**Architectural Observation for your Notes:** Notice how dynamically the questions shifted. Because a hybrid TAM/Architect role balances client communication with deep tech stacks, the parallel agents automatically isolated their domains: Gemini forces you to defend your executive relationship skills, Cerebras tests if you actually know your GCP infrastructure under the certification layer, and OpenAI probes how you handle enterprise security risk without abdicating operational responsibility.
"""

display(Markdown(panel_report))
````
