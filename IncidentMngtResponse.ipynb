```python
import os
import pandas as pd
from openai import OpenAI
from IPython.display import display

# Ensure Pandas displays full text lines in the notebook
pd.set_option("display.max_colwidth", None)

# Initialize clients (Unified OpenAI protocol)
openai_client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
gemini_client = OpenAI(
    api_key=os.getenv("GOOGLE_API_KEY"),
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)
cerebras_client = OpenAI(
    api_key=os.getenv("CEREBRAS_API_KEY"),
    base_url="https://api.cerebras.ai/v1"
)

# =====================================================================
# 1. DEFINE THE COMMON DDOS SECURITY INCIDENT INPUT
# =====================================================================
security_incident = """
CRITICAL SECURITY ALERT: ACTIVE DDOS ATTACK
ATTACK TYPE: Volumetric Layer 7 HTTP Flood & Syn Flood
CURRENT IMPACT: 
The primary customer login portal is experiencing a 400% spike in malicious 
traffic, originating from a highly distributed global botnet. 
BUSINESS CONTEXT:
The authentication servers are running at 98% CPU utilization. Legit users 
are experiencing 15-second page load times or outright connection drops. 
Turning on maximum Cloudflare/WAF 'Under Attack' mode will filter out the botnet 
but will challenge 100% of legitimate users with aggressive CAPTCHAs, 
severely lowering checkout conversion rates during a major promotional sale.
"""

# =====================================================================
# 2. ARCHITECTURE 1: PARALLELIZATION (VOTING/SECTIONAL)
# =====================================================================
print("--- Executing Architecture 1: Parallelization ---")

user_prompt = f"Analyze this vulnerability and issue your primary defense action:\n\n{security_incident}"

# Run independent expert personas completely blind to each other
print("Polling OpenAI (Defensive Security Engineer)...")
op_parallel = openai_client.chat.completions.create(
    model="gpt-5.4-nano",
    messages=[{"role": "system", "content": "You are a Defensive Security Engineer. Focus purely on immediate containment, firewalls, and network isolation."},
              {"role": "user", "content": user_prompt}]
)

print("Polling Gemini (Business Continuity Director)...")
gem_parallel = gemini_client.chat.completions.create(
    model="gemini-2.5-flash-lite",
    messages=[{"role": "system", "content": "You are a Business Continuity Director. Focus on minimizing user downtime and protecting transactional revenue lines."},
              {"role": "user", "content": user_prompt}]
)

print("Polling Cerebras (Corporate Compliance Officer)...")
cer_parallel = cerebras_client.chat.completions.create(
    model="gpt-oss-120b",
    messages=[{"role": "system", "content": "You are a Corporate Compliance and Legal Officer. Focus strictly on SLA breach liabilities regarding platform availability metrics, data privacy handling during traffic redirection, and managing executive crisis notifications for public stakeholders."},
              {"role": "user", "content": user_prompt}]
)

# =====================================================================
# 3. ORGANIZE AND DISPLAY PARALLEL MATRIX VIEW (Zero-Installation HTML)
# =====================================================================
parallel_data = {
    "Architecture: Parallelization": ["Strategic Response Viewpoint"],
    "OpenAI (Security Engine)": [op_parallel.choices[0].message.content],
    "Gemini (Continuity Expert)": [gem_parallel.choices[0].message.content],
    "Cerebras (Compliance Officer)": [cer_parallel.choices[0].message.content]
}

df_parallel = pd.DataFrame(parallel_data)
df_parallel.set_index("Architecture: Parallelization", inplace=True)

print("\n" + "="*50)
print("PARALLEL EVALUATION MATRIX DISPLAY")
print("="*50)
display(df_parallel)
```

```python
import os
import pandas as pd
from openai import OpenAI
from IPython.display import display, Markdown  # 1. FIXED: Added native markdown render engine

# Ensure clean global string tracking
pd.set_option("display.max_colwidth", None)

# Initialize endpoints using the unified API protocol
openai_client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
gemini_client = OpenAI(
    api_key=os.getenv("GEMINI_API_KEY"),
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)
cerebras_client = OpenAI(
    api_key=os.getenv("CEREBRAS_API_KEY"),
    base_url="https://api.cerebras.ai/v1"
)
```

```python
# =====================================================================
# 1. DEFINE THE SCENARIO (FIXED: Isolated Context Block)
# =====================================================================
security_incident = """
CRITICAL SECURITY ALERT: ACTIVE DDOS ATTACK
ATTACK TYPE: Volumetric Layer 7 HTTP Flood & Syn Flood
CURRENT IMPACT: 
The primary customer login portal is experiencing a 400% spike in malicious 
traffic, originating from a highly distributed global botnet. 
BUSINESS CONTEXT:
The authentication servers are running at 98% CPU utilization. Legit users 
are experiencing 15-second page load times or outright connection drops. 
Turning on maximum Cloudflare/WAF 'Under Attack' mode will filter out the botnet 
but will challenge 100% of legitimate users with aggressive CAPTCHAs, 
severely lowering checkout conversion rates during a major promotional sale.
"""

# =====================================================================
# 2. ARCHITECTURE 3: EVALUATOR-OPTIMIZER (The Multi-Turn Loop)
# =====================================================================
print("--- Executing Architecture 3: Evaluator-Optimizer (DDoS Theme) ---\n")

# Step 1: Generator drafts the initial response statement to the public
generator_prompt = f"Draft a public relations statement explaining why our login portal is experiencing latency issues: {security_incident}"

print("Generating Initial Draft (Cerebras)...")
gen_res = cerebras_client.chat.completions.create(
    model="gpt-oss-120b",
    messages=[
        {
            "role": "system", 
            "content": "You are a corporate PR Copywriter. CRITICAL REQUIREMENT: Your output must be EXACTLY two sentences. Keep it brief, professional, and clear."
        },
        {"role": "user", "content": generator_prompt}
    ],
)
current_draft = gen_res.choices[0].message.content
print(f"-> Initial Draft Created.\n")
```

```python
# Step 2: Create the multi-turn evaluation loop
for turn in range(1, 3):
    print(f"🔄 Starting Optimization Turn {turn}...")
    
    # Evaluator reviews the draft
    eval_prompt = f"Review this PR draft. Does it leak technical secrets (like mentioning HTTP/Syn floods or CPU metrics) or cause unnecessary panic? Draft: {current_draft}"
    
    eval_res = openai_client.chat.completions.create(
        model="gpt-5.4-nano",
        messages=[
            {
                "role": "system", 
                "content": "You are a Senior Security Legal Auditor. Critique the draft. If it contains technical leaks or panic phrasing, give explicit, concise correction feedback. If it is 100% safe, professional, and hides technical details, reply with ONLY the word 'APPROVED'."
            },
            {"role": "user", "content": eval_prompt}
        ],
        temperature=0.1
    )
    feedback = eval_res.choices[0].message.content
    print(f"   [Evaluator Verdict]: {feedback}\n")
    
    # Check if the safety gate is successfully cleared
    if "APPROVED" in feedback.upper():
        print("✅ Draft successfully passed the safety gate early! Ending loop.")
        break

# Generator optimizes based on feedback
    optimizer_prompt = f"Rewrite your original draft to fully fix the legal feedback. Original: {current_draft}\nFeedback: {feedback}"
    
    gen_res = cerebras_client.chat.completions.create(
        model="gpt-oss-120b",
        messages=[
            {
                "role": "system", 
                "content": "You are a corporate PR Copywriter. Rewrite the statement to fix the audit feedback. CRITICAL CONSTRAINTS: Your response must remain EXACTLY two sentences long. No intro fluff, no explanations."
            },
            {"role": "user", "content": optimizer_prompt}
        ],
    )
    current_draft = gen_res.choices[0].message.content
    print(f"   [Optimized Draft {turn}]: {current_draft}\n")


```

```python
# =====================================================================
# 3. RENDER THE FINAL POLISHED PR BROADCAST (Stacked Markdown Layout)
# =====================================================================
print("\n" + "="*50)
print("FINAL EVALUATOR-OPTIMIZER OUTCOME")
print("="*50)

final_report = f"""
# 📢 Public Relations Incident Broadcast
**Architecture Pattern:** Evaluator-Optimizer (Multi-Turn Quality Gate)

---

### 🛡️ Final Sanatized Output (Approved for Public Use)
> {current_draft}

---
"""
display(Markdown(final_report))
```
