Architecture Pattern 1: Parallelization (Section / Voting)
Anthropic Definition: Running the same task across multiple diverse models or personas simultaneously to get different viewpoints or to vote on a consensus.

When to use: Deep risk auditing or brainstorming when you want to avoid single-model bias.

```python
import os
import pandas as pd
from openai import OpenAI
from IPython.display import display

# Ensure Pandas displays full text lines in the notebook
pd.set_option("display.max_colwidth", None)

# Initialize clients (Unified OpenAI protocol)
openai_client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
gemini_client = OpenAI(
    api_key=os.getenv("GOOGLE_API_KEY"),
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)
cerebras_client = OpenAI(
    api_key=os.getenv("CEREBRAS_API_KEY"),
    base_url="https://api.cerebras.ai/v1"
)

# =====================================================================
# 1. DEFINE THE COMMON DDOS SECURITY INCIDENT INPUT
# =====================================================================
security_incident = """
CRITICAL SECURITY ALERT: ACTIVE DDOS ATTACK
ATTACK TYPE: Volumetric Layer 7 HTTP Flood & Syn Flood
CURRENT IMPACT: 
The primary customer login portal is experiencing a 400% spike in malicious 
traffic, originating from a highly distributed global botnet. 
BUSINESS CONTEXT:
The authentication servers are running at 98% CPU utilization. Legit users 
are experiencing 15-second page load times or outright connection drops. 
Turning on maximum Cloudflare/WAF 'Under Attack' mode will filter out the botnet 
but will challenge 100% of legitimate users with aggressive CAPTCHAs, 
severely lowering checkout conversion rates during a major promotional sale.
"""

# =====================================================================
# 2. ARCHITECTURE 1: PARALLELIZATION (VOTING/SECTIONAL)
# =====================================================================
print("--- Executing Architecture 1: Parallelization ---")

user_prompt = f"Analyze this vulnerability and issue your primary defense action:\n\n{security_incident}"

# Run independent expert personas completely blind to each other
print("Polling OpenAI (Defensive Security Engineer)...")
op_parallel = openai_client.chat.completions.create(
    model="gpt-4o",
    messages=[{"role": "system", "content": "You are a Defensive Security Engineer. Focus purely on immediate containment, firewalls, and network isolation."},
              {"role": "user", "content": user_prompt}]
)

print("Polling Gemini (Business Continuity Director)...")
gem_parallel = gemini_client.chat.completions.create(
    model="gemini-2.5-flash-lite",
    messages=[{"role": "system", "content": "You are a Business Continuity Director. Focus on minimizing user downtime and protecting transactional revenue lines."},
              {"role": "user", "content": user_prompt}]
)

print("Polling Cerebras (Corporate Compliance Officer)...")
cer_parallel = cerebras_client.chat.completions.create(
    model="gpt-oss-120b",
    messages=[{"role": "system", "content": "You are a Corporate Compliance and Legal Officer. Focus strictly on SLA breach liabilities regarding platform availability metrics, data privacy handling during traffic redirection, and managing executive crisis notifications for public stakeholders."},
              {"role": "user", "content": user_prompt}]
)

# =====================================================================
# 3. ORGANIZE AND DISPLAY PARALLEL MATRIX VIEW (Zero-Installation HTML)
# =====================================================================
parallel_data = {
    "Architecture: Parallelization": ["Strategic Response Viewpoint"],
    "OpenAI (Security Engine)": [op_parallel.choices[0].message.content],
    "Gemini (Continuity Expert)": [gem_parallel.choices[0].message.content],
    "Cerebras (Compliance Officer)": [cer_parallel.choices[0].message.content]
}

df_parallel = pd.DataFrame(parallel_data)
df_parallel.set_index("Architecture: Parallelization", inplace=True)

print("\n" + "="*50)
print("PARALLEL EVALUATION MATRIX DISPLAY")
print("="*50)
display(df_parallel)
```

Architecture Pattern 2: Orchestrator-Workers
Anthropic Definition: A central coordinator model breaks down a complex task dynamically into unique sub-tasks, delegates them to specialized workers, and synthesizes the outputs.

When to use: Open-ended tasks where you cannot predict what actions are required upfront.

```python

import json  # <-- Add this right at the top!
import os
import pandas as pd
from openai import OpenAI
from IPython.display import display

# Your existing clients and variable initializations...
openai_client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

# (The rest of your Orchestrator-Workers code follows below perfectly)

# =====================================================================
# ARCHITECTURE 2: ORCHESTRATOR-WORKERS
# =====================================================================
print("\n--- Executing Architecture 2: Orchestrator-Workers ---")

# Step 1: Orchestrator analyzes the incident and creates the specialized work assignments
orchestrator_prompt = f"""
Analyze this security crisis and break it down into exactly two distinct operational work assignments.
Return your answer in a strict JSON object with the keys 'task_for_worker_1' and 'task_for_worker_2'.
Incident Details:
{security_incident}
"""

orch_res = openai_client.chat.completions.create(
    model="gpt-4o",
    messages=[{"role": "system", "content": "You are the Incident Commander. Break down tasks logically."},
              {"role": "user", "content": orchestrator_prompt}],
    response_format={"type": "json_object"}
)
tasks = json.loads(orch_res.choices[0].message.content)
```

```python
# Step 2: Dynamically pass the orchestrator's unique tasks to specialized workers
print(f" -> Deploying Worker 1 (Gemini) on Infrastructure Mitigation...")
worker_1_res = gemini_client.chat.completions.create(
    model="gemini-2.5-flash-lite",
    messages=[{"role": "system", "content": "You are a Network Engineering Worker. Provide an explicit, step-by-step mitigation plan for the assigned task."},
              {"role": "user", "content": tasks['task_for_worker_1']}]
)

print(f" -> Deploying Worker 2 (Cerebras) on Comms and Revenue Protection...")
# Keeping Cerebras short and direct using system prompt constraints
worker_2_res = cerebras_client.chat.completions.create(
    model="gpt-oss-120b",
    messages=[{"role": "system", "content": "You are a Customer Communications Worker. Provide an explicit, 2-sentence action plan for the assigned task. Get straight to the point without any introductory fluff."},
              {"role": "user", "content": tasks['task_for_worker_2']}],
    temperature=0.1,
    max_tokens=150
)

print("\nAll workers completed their dynamic assignments.")
```

```python
from IPython.display import display, Markdown  # <-- Add Markdown here!
# =====================================================================
# 3. ORGANIZE AND DISPLAY THE ORCHESTRATION BRIEFING (Clean Markdown Layout)
# =====================================================================
print("\nFormatting final execution report...")

# Extract strings cleanly from our runtime variables
orchestrator_task_1 = tasks['task_for_worker_1']
orchestrator_task_2 = tasks['task_for_worker_2']
worker_1_output = worker_1_res.choices[0].message.content
worker_2_output = worker_2_res.choices[0].message.content

# Construct a clean, vertically stacked Markdown document
briefing_output = f"""
# 📑 Incident Command Orchestration Report
**Incident Type:** Active Layer 7 & Syn Flood DDoS Attack  
**Orchestration Pattern:** Orchestrator-Workers (Dynamic Task Delegation)

---

## 🧠 1. Incident Commander Tactical Breakdown (Orchestrator)
The central orchestrator analyzed the active 98% CPU load and divided the crisis response operations into two parallel workstreams:

### ⚙️ Workstream Alpha: Infrastructure Mitigation
* **Assigned Worker:** Gemini (Network Operations Specialist)
* **Dynamic Directive Issued:** > {orchestrator_task_1}

### 📢 Workstream Bravo: Business Communications
* **Assigned Worker:** Cerebras (Customer Experience Specialist)
* **Dynamic Directive Issued:** > {orchestrator_task_2}

---

## 🛠️ 2. Parallel Worker Execution Logs

### 🔴 Worker 1 Response (Gemini — Infrastructure Engineering)
{worker_1_output}

### 🔵 Worker 2 Response (Cerebras — Communications & Revenue)
{worker_2_output}

---
**End of Briefing. All mitigation tracks active.**
"""

# Render the entire document beautifully inside your notebook cell
print("\n" + "="*50)
print("DYNAMIC ARCHITECTURE EXECUTION BRIEFING")
print("="*50)
display(Markdown(briefing_output))
```

Architecture Pattern 3: Evaluator-Optimizer
Anthropic Definition: A multi-turn feedback loop. One model acts as the generator to draft a solution, while another acts as an evaluator to critique it, sending it back for optimization until it passes a high-quality gate.

When to use: When you have a clear rubric for success and cannot afford any errors (e.g., code compilation, public relations statements).

[ Input: DDoS Incident Data ]
             │
             ▼
┌────────────────────────────────────────┐
│  STEP 1: GENERATOR (Gemini 1.5 Pro)    │ ◄────────────────┐
│  Drafts/Rewrites a 2-Sentence PR Memo  │                  │
└───────────────────┬────────────────────┘                  │
                    │                                       │
                    ▼ (Current Draft String)                │
┌────────────────────────────────────────┐                  │
│  STEP 2: EVALUATOR (OpenAI gpt-4o)     │                  │
│  Audits draft for leaks or panic words │                  │
└───────────────────┬────────────────────┘                  │
                    │                                       │
                    ├── [ If leaks found: Sends Feedback ]──┘
                    │
                    └── [ If Safe: Replies "APPROVED" ]
                             │
                             ▼
                 [ Production Broadcast ]

```python
import os
import pandas as pd
from openai import OpenAI
from IPython.display import display, Markdown  # 1. FIXED: Added native markdown render engine

# Ensure clean global string tracking
pd.set_option("display.max_colwidth", None)

# Initialize endpoints using the unified API protocol
openai_client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
gemini_client = OpenAI(
    api_key=os.getenv("GEMINI_API_KEY"),
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)
cerebras_client = OpenAI(
    api_key=os.getenv("CEREBRAS_API_KEY"),
    base_url="https://api.cerebras.ai/v1"
)
```

```python
# =====================================================================
# 1. DEFINE THE SCENARIO (FIXED: Isolated Context Block)
# =====================================================================
security_incident = """
CRITICAL SECURITY ALERT: ACTIVE DDOS ATTACK
ATTACK TYPE: Volumetric Layer 7 HTTP Flood & Syn Flood
CURRENT IMPACT: 
The primary customer login portal is experiencing a 400% spike in malicious 
traffic, originating from a highly distributed global botnet. 
BUSINESS CONTEXT:
The authentication servers are running at 98% CPU utilization. Legit users 
are experiencing 15-second page load times or outright connection drops. 
Turning on maximum Cloudflare/WAF 'Under Attack' mode will filter out the botnet 
but will challenge 100% of legitimate users with aggressive CAPTCHAs, 
severely lowering checkout conversion rates during a major promotional sale.
"""

# =====================================================================
# 2. ARCHITECTURE 3: EVALUATOR-OPTIMIZER (The Multi-Turn Loop)
# =====================================================================
print("--- Executing Architecture 3: Evaluator-Optimizer (DDoS Theme) ---\n")

# Step 1: Generator drafts the initial response statement to the public
generator_prompt = f"Draft a public relations statement explaining why our login portal is experiencing latency issues: {security_incident}"

print("Generating Initial Draft (Cerebras)...")
gen_res = cerebras_client.chat.completions.create(
    model="gpt-oss-120b",
    messages=[
        {
            "role": "system", 
            "content": "You are a corporate PR Copywriter. CRITICAL REQUIREMENT: Your output must be EXACTLY two sentences. Keep it brief, professional, and clear."
        },
        {"role": "user", "content": generator_prompt}
    ],
)
current_draft = gen_res.choices[0].message.content
print(f"-> Initial Draft Created.\n")
```

```python
# Step 2: Create the multi-turn evaluation loop
for turn in range(1, 3):
    print(f"🔄 Starting Optimization Turn {turn}...")
    
    # Evaluator reviews the draft
    eval_prompt = f"Review this PR draft. Does it leak technical secrets (like mentioning HTTP/Syn floods or CPU metrics) or cause unnecessary panic? Draft: {current_draft}"
    
    eval_res = openai_client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {
                "role": "system", 
                "content": "You are a Senior Security Legal Auditor. Critique the draft. If it contains technical leaks or panic phrasing, give explicit, concise correction feedback. If it is 100% safe, professional, and hides technical details, reply with ONLY the word 'APPROVED'."
            },
            {"role": "user", "content": eval_prompt}
        ],
        temperature=0.1
    )
    feedback = eval_res.choices[0].message.content
    print(f"   [Evaluator Verdict]: {feedback}\n")
    
    # Check if the safety gate is successfully cleared
    if "APPROVED" in feedback.upper():
        print("✅ Draft successfully passed the safety gate early! Ending loop.")
        break

# Generator optimizes based on feedback
    optimizer_prompt = f"Rewrite your original draft to fully fix the legal feedback. Original: {current_draft}\nFeedback: {feedback}"
    
    gen_res = cerebras_client.chat.completions.create(
        model="gpt-oss-120b",
        messages=[
            {
                "role": "system", 
                "content": "You are a corporate PR Copywriter. Rewrite the statement to fix the audit feedback. CRITICAL CONSTRAINTS: Your response must remain EXACTLY two sentences long. No intro fluff, no explanations."
            },
            {"role": "user", "content": optimizer_prompt}
        ],
    )
    current_draft = gen_res.choices[0].message.content
    print(f"   [Optimized Draft {turn}]: {current_draft}\n")


```

```python
# =====================================================================
# 3. RENDER THE FINAL POLISHED PR BROADCAST (Stacked Markdown Layout)
# =====================================================================
print("\n" + "="*50)
print("FINAL EVALUATOR-OPTIMIZER OUTCOME")
print("="*50)

final_report = f"""
# 📢 Public Relations Incident Broadcast
**Architecture Pattern:** Evaluator-Optimizer (Multi-Turn Quality Gate)

---

### 🛡️ Final Sanatized Output (Approved for Public Use)
> {current_draft}

---
### 📊 Architectural Insights for your Lab Report:
* **The Evaluator (OpenAI gpt-4o):** Acted as a strict quality filter, catching technical vulnerabilities or security jargon (like 'Layer 7 HTTP Flood') before the public can see it.
* **The Optimizer Loop:** Showcases how a model can self-correct its own vocabulary over multiple iterations without human developer intervention.
"""

display(Markdown(final_report))
```
