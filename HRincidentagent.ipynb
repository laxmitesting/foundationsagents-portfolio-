```python
import os
import json
from openai import OpenAI
from dotenv import load_dotenv
```

```python
load_dotenv(override=True)
```

```python

# 1. Standard OpenAI Client
openai_client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY")
)

# 2. Gemini Client (routed via Google's OpenAI-compatibility endpoint)
gemini_client = OpenAI(
    api_key=os.getenv("GOOGLE_API_KEY"),
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)

# 3. Cerebras Client (routed via Cerebras's OpenAI-compatibility endpoint)
cerebras_client = OpenAI(
    api_key=os.getenv("CEREBRAS_API_KEY"),
    base_url="https://api.cerebras.ai/v1"
)
```

```python
# =====================================================================
# THE REPORT TEXT
# =====================================================================
# Copy this directly into your existing notebook cell
hr_incident_report = """
INCIDENT TRACKING ID: HR-2026-9941
DATE OF REPORT: May 29, 2026
SUBMITTED BY: Anonymous Escalation (via Secure Employee Portal)
DEPARTMENT: Global Infrastructure Engineering

INCIDENT SUMMARY:
During the Q2 Architecture Review session on Wednesday, a Senior Team Lead publicly reprimanded a junior cloud engineer. Multiple witnesses reported that the Lead cut off the junior employee mid-sentence, stating: 'Your architectural proposals are a fundamental waste of time. If you cannot comprehend basic Kubernetes networking concepts, maybe you should transfer to manual QA testing instead of core system design.'

IMPACT & CONSEQUENCES:
The junior engineer immediately muted their microphone, turned off their camera, and left the video conference abruptly. They have been marked as 'Out of Office' and unresponsive to direct team pings for the last 48 hours. 

HISTORICAL CONTEXT:
This is the third formal flag regarding this specific Senior Team Lead's abrasive communication style and exclusionary micro-aggressions this quarter. Prior verbal warnings were issued by the Engineering Director on March 12th regarding psychological safety boundaries.
"""

user_prompt = f"Analyze the following internal HR incident report:\n\n{hr_incident_report}"
agent_responses = {}
```

```python

# Call OpenAI
op_res = openai_client.chat.completions.create(
    model="gpt-4o",
    messages=[{"role": "system", "content": "You are a Legal Auditor."}, 
              {"role": "user", "content": user_prompt}]
)
agent_responses["OpenAI"] = op_res.choices[0].message.content

print(op_res.choices[0].message.content)
```

```python
# Call Gemini 
gem_res = gemini_client.chat.completions.create(
    model="gemini-2.5-flash", 
    messages=[{"role": "system", "content": "You are an Industrial Psychologist."}, 
              {"role": "user", "content": user_prompt}]
)
agent_responses["Gemini"] = gem_res.choices[0].message.content

print(gem_res.choices[0].message.content)
```

```python
# Call Cerebras 
cer_res = cerebras_client.chat.completions.create(
    model="gpt-oss-120b",
    messages=[{"role": "system", "content": "You are an HR Ops Specialist."}, 
              {"role": "user", "content": user_prompt}]
)
agent_responses["Cerebras"] = cer_res.choices[0].message.content

print(cer_res.choices[0].message.content)
```
